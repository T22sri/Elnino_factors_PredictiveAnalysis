```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
# Loading and Manipulate Data

#https://drive.google.com/drive/folders/1H0-0943BbMWFpgM6fCDtCqmQnGi7WAjq?usp=sharing
df_nino = pd.read_csv('/content/gdrive/MyDrive/603-MachineLearning/Project/elnino.csv')

# Remove extra space in columns
df_nino.columns = [col.strip() for col in df_nino.columns]

df_nino.head()
df_nino
```

```python
# Informations of Elnino DataFrame

print('Informations of DataFrame:','\n')
print(df_nino.info(),'\n')

print('-'*100,'\n')

print('Description of DataFrame:','\n')
print(df_nino.describe())

print('-'*100,'\n')

print('Name of columns:', '\n')
print(df_nino.columns) #Have Spaces before the name of some columns
```

```python
# Naming features
names = list(df_nino.columns)

for i in range(0,12):
    names[i] = names[i].lstrip()

df_nino.columns = names

# Remove missing values
df_nino.replace('.',np.nan,inplace = True)
df_nino
```

```python
df_nino.describe(include='all')
```

```python
df_nino.info()
```

```python
# To convert to Float64 types

for column in df_nino[['Zonal Winds', 'Meridional Winds', 'Humidity', 'Air Temp', 'Sea Surface Temp']]:
    df_nino[column] = df_nino[column].astype("float64")

df_nino.dtypes
```

```python
# Separate numerical (or main) features
numerical_features=list(df_nino.columns)
numerical_features.remove('Observation')
numerical_features.remove('Year')
numerical_features.remove('Month')
numerical_features.remove('Day')
numerical_features.remove('Date')
print(numerical_features)
```

```python
# Scatter plot
f,ax = plt.subplots(2,3,figsize=(11,9))

sns.scatterplot('Latitude', 'Sea Surface Temp', data=df_nino, color='g', ax=ax[0,0])
sns.scatterplot('Longitude', 'Sea Surface Temp', data=df_nino, color='b', ax=ax[0,1])
sns.scatterplot('Zonal Winds', 'Sea Surface Temp', data=df_nino, color='r', ax=ax[0,2])
sns.scatterplot('Meridional Winds', 'Sea Surface Temp', data=df_nino, color='orange', ax=ax[1,0])
sns.scatterplot('Humidity', 'Sea Surface Temp', data=df_nino, color='purple', ax=ax[1,1])
sns.scatterplot('Air Temp', 'Sea Surface Temp', data=df_nino, color='yellow', ax=ax[1,2])

ax[0,1].set_ylabel('')
ax[0,2].set_ylabel('')
ax[1,1].set_ylabel('')
ax[1,2].set_ylabel('')
```

```python
# Normality test on Sea Surface Temp
from scipy.stats import normaltest

stat, p = normaltest(df_nino['Sea Surface Temp'])
print('Statistics=%.3f, p=%.3f' % (stat, p))

alpha = 0.05
if p < alpha:  # null hypothesis: x comes from a normal distribution
    print("The null hypothesis can be rejected")
else:
    print("The null hypothesis cannot be rejected")
```

```python
# Shapiro-Wilk Test on Sea Surface Temp 
from scipy.stats import shapiro

stat, p = shapiro(df_nino['Sea Surface Temp'])
print('Statistics=%.3f, p=%.3f' % (stat, p))

if p > 0.05:  
    print("Probably Gaussian")
else:
    print("Probably not Gaussian")
```
